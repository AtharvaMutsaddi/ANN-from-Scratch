# Value Scaling

```python
def std_scaling(df, column):
    mu = df[column].mean()
    sigma = df[column].std()
    df[column] = (df[column] - mu) / sigma
    return df[column]
```

```python
columns=df.columns
ignore_cols=['id','defects']
for col in columns:
    if col not in ignore_cols:
        df[col]=std_scaling(df,col)
```

# Train Test Split

```python
def train_test_split(df,train_percentage):
    np.random.seed(1412)
    num_rows = len(df)
    num_train = int(train_percentage * num_rows)
    # Generate random indices for training and testing
    indices = np.random.permutation(num_rows)
    # Training set indices
    train_indices = indices[:num_train]
    # Testing set indices
    test_indices = indices[num_train:]

    # Create training and testing sets
    x_train = df.iloc[train_indices].drop(columns=['defects','id']).values
    y_train = df.iloc[train_indices]['defects'].values.reshape(-1, 1)

    x_test = df.iloc[test_indices].drop(columns=['defects','id']).values
    y_test = df.iloc[test_indices]['defects'].values.reshape(-1, 1)
    return x_train,x_test,y_train,y_test
```

```python
class Layer_Dense():
    def __init__(self, n_inputs, n_neurons):
        lower_lim=(-1*(2/(n_inputs+n_neurons))**0.5)
        upper_lim=((2/(n_inputs+n_neurons))**0.5)
        self.weights=np.random.uniform(lower_lim,upper_lim,size=(n_inputs, n_neurons))
        self.biases = np.zeros((1, n_neurons))
        self.inputs = None
        self.dweights = None
        self.dbiases = None
        self.dinputs = None

    def forward(self, inputs):
        self.inputs = inputs
        self.output = np.dot(inputs, self.weights) + self.biases

    def backward(self, dvalues):
        self.dweights = np.dot(self.inputs.T, dvalues)
        self.dbiases = np.sum(dvalues, axis=0, keepdims=True)
        self.dinputs = np.dot(dvalues, self.weights.T)

    def clip_gradients(self, threshold):
        self.dweights = np.clip(self.dweights, -threshold, threshold)
        self.dbiases = np.clip(self.dbiases, -threshold, threshold)
```

```python
class Activation_Sigmoid:
    def forward(self,inputs):
        self.output = 1/(1+np.exp(-inputs))
        
    def backward(self,dvalues):
        sigmoid_derivative = self.output*(1-self.output)
        self.dinputs = dvalues*sigmoid_derivative
```

```python
class Activation_ReLU:
    
    def forward(self, inputs):
        self.output = np.maximum(inputs,0)
    def backward(self,dvalues):
        self.dinputs=dvalues.copy()
        self.dinputs[self.dinputs<=0]=0
```

```python

class Loss_MSE:
    def forward(self,y_pred,y_true):
        sample_losses=np.mean((y_true-y_pred)**2,axis=-1)
        return sample_losses
    def backward(self,dvalues,y_true):
        samples=len(dvalues)
        outputs=len(dvalues[0])
        self.dinputs=(-2*(y_true-dvalues)/outputs)/samples


        
```

```python
class Loss_BinaryCrossEntropy:
    def __init__(self, epsilon=1e-15):
        self.epsilon = epsilon
        self.dinputs = None
    def forward(self, y_pred, y_true):
        y_pred = np.clip(y_pred, self.epsilon, 1 - self.epsilon)
        sample_losses = -(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
        return np.mean(sample_losses, axis=-1)
    def backward(self, dvalues, y_true):
        outputs = len(dvalues[0])
        self.dinputs = (-(y_true / dvalues - (1 - y_true) / (1 - dvalues)))
        self.dinputs = self.dinputs / len(dvalues)
```

# Evaluation Process

```python
def convt_to_binary(sigmoid_predictions):
    for ele in sigmoid_predictions:
        if ele[0] >0.5:
            ele[0]=int(1)
        else:
            ele[0]=int(0)

    return sigmoid_predictions
```

```python
def calculate_metrics(y_true, y_pred):
    sigmoid_predictions=y_pred.copy()
    sigmoid_predictions=convt_to_binary(sigmoid_predictions).astype(int)
    # Accuracy
    accuracy = np.mean(sigmoid_predictions == y_true)

    # True Positives, True Negatives, False Positives, False Negatives
    tp = np.sum((sigmoid_predictions == 1) & (y_true == 1))
    fp = np.sum((sigmoid_predictions == 1) & (y_true == 0))
    fn = np.sum((sigmoid_predictions == 0) & (y_true == 1))

    # Precision, Recall, and F1 Score
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    return accuracy, precision, recall, f1_score
```

```python
layer1 = Layer_Dense(21, 200)
activation1 = Activation_ReLU()

layer2 = Layer_Dense(200, 100)
activation2 = Activation_ReLU()

layer3 = Layer_Dense(100, 1)
activation3 = Activation_Sigmoid()
```

# Training Using MSE

```python
loss_function = Loss_MSE()
learning_rate = 0.003
gradient_clip_threshold = 1
epochs = 1000
```

```python
acc=[]
prec=[]
recalls=[]
f1=[]
plot_epochs=[]
for epoch in range(epochs):
    layer1.forward(x_train)
    activation1.forward(layer1.output)

    layer2.forward(activation1.output)
    activation2.forward(layer2.output)

    layer3.forward(activation2.output)
    activation3.forward(layer3.output)
    loss = loss_function.forward(activation3.output, y_train)
    if np.isnan(loss).any() or np.isinf(loss).any():
        print(f'NaN or Inf values encountered in loss. Stopping training.')
        break

    loss_function.backward(activation3.output, y_train)
    
    activation3.backward(loss_function.dinputs)
    layer3.backward(activation3.dinputs)
    activation2.backward(layer3.dinputs)
    layer2.backward(activation2.dinputs)
    activation1.backward(layer2.dinputs)
    layer1.backward(activation1.dinputs)

    layer1.clip_gradients(gradient_clip_threshold)
    layer2.clip_gradients(gradient_clip_threshold)
    layer3.clip_gradients(gradient_clip_threshold)

    if (
        np.isnan(layer1.dweights).any() or np.isinf(layer1.dweights).any() or
        np.isnan(layer1.dbiases).any() or np.isinf(layer1.dbiases).any() or
        np.isnan(layer2.dweights).any() or np.isinf(layer2.dweights).any() or
        np.isnan(layer2.dbiases).any() or np.isinf(layer2.dbiases).any() or
        np.isnan(layer3.dweights).any() or np.isinf(layer3.dweights).any() or
        np.isnan(layer3.dbiases).any() or np.isinf(layer3.dbiases).any()
    ):
        print(f'NaN or Inf values encountered in gradients. Stopping training.')
        break

    layer1.weights += -learning_rate * layer1.dweights
    layer1.biases += -learning_rate * layer1.dbiases

    layer2.weights += -learning_rate * layer2.dweights
    layer2.biases += -learning_rate * layer2.dbiases

    layer3.weights += -learning_rate * layer3.dweights
    layer3.biases += -learning_rate * layer3.dbiases
    if epoch%10==0:
        accuracy, precision, recall, f1_score=calculate_metrics(y_train,activation3.output)
        print(f"Epoch:{epoch}. Accuracy:{accuracy*100},Precision:{precision*100},Recall:{recall*100},F1 Score:{f1_score*100}")
        plot_epochs.append(epoch)
        acc.append(accuracy)
        prec.append(precision)
        recalls.append(recall)
        f1.append(f1_score)
```

```python
predictions = activation3.output
accuracy, precision, recall, f1_score=calculate_metrics(y_train,predictions)
print("Metrics on the Training Data")
print(f"Accuracy:{accuracy*100}%")
print(f"Precision:{precision*100}%")
print(f"Recall:{recall*100}%")
print(f"F1 Score:{f1_score*100}%")
```

```python
plt.plot(plot_epochs,acc)
plt.plot(plot_epochs,f1)
plt.plot(plot_epochs,recalls)
plt.plot(plot_epochs,prec)
plt.legend(['accuracy','f1','recall','precision'])
plt.show()
```

# Trying with Different Epochs and Premature Stopping

```python
layer1 = Layer_Dense(21, 200)
activation1 = Activation_ReLU()

layer2 = Layer_Dense(200, 100)
activation2 = Activation_ReLU()

layer3 = Layer_Dense(100, 1)
activation3 = Activation_Sigmoid()
```

```python
loss_function = Loss_BinaryCrossEntropy()
learning_rate = 0.002
gradient_clip_threshold = 1
epochs = 1250
```

```python
acc=[]
prec=[]
recalls=[]
f1=[]
plot_epochs=[]
det_cnt=0
for epoch in range(epochs):
    layer1.forward(x_train)
    activation1.forward(layer1.output)

    layer2.forward(activation1.output)
    activation2.forward(layer2.output)

    layer3.forward(activation2.output)
    activation3.forward(layer3.output)
    loss = loss_function.forward(activation3.output, y_train)
    if np.isnan(loss).any() or np.isinf(loss).any():
        print(f'NaN or Inf values encountered in loss. Stopping training.')
        break

    loss_function.backward(activation3.output, y_train)
    
    activation3.backward(loss_function.dinputs)
    layer3.backward(activation3.dinputs)
    activation2.backward(layer3.dinputs)
    layer2.backward(activation2.dinputs)
    activation1.backward(layer2.dinputs)
    layer1.backward(activation1.dinputs)

    layer1.clip_gradients(gradient_clip_threshold)
    layer2.clip_gradients(gradient_clip_threshold)
    layer3.clip_gradients(gradient_clip_threshold)

    if (
        np.isnan(layer1.dweights).any() or np.isinf(layer1.dweights).any() or
        np.isnan(layer1.dbiases).any() or np.isinf(layer1.dbiases).any() or
        np.isnan(layer2.dweights).any() or np.isinf(layer2.dweights).any() or
        np.isnan(layer2.dbiases).any() or np.isinf(layer2.dbiases).any() or
        np.isnan(layer3.dweights).any() or np.isinf(layer3.dweights).any() or
        np.isnan(layer3.dbiases).any() or np.isinf(layer3.dbiases).any()
    ):
        print(f'NaN or Inf values encountered in gradients. Stopping training.')
        break

    layer1.weights += -learning_rate * layer1.dweights
    layer1.biases += -learning_rate * layer1.dbiases

    layer2.weights += -learning_rate * layer2.dweights
    layer2.biases += -learning_rate * layer2.dbiases

    layer3.weights += -learning_rate * layer3.dweights
    layer3.biases += -learning_rate * layer3.dbiases
    if epoch%10==0:
        accuracy, precision, recall, f1_score=calculate_metrics(y_train,activation3.output)
        if len(acc)>0:
            if accuracy<acc[len(acc)-1]:
                det_cnt+=1
                if det_cnt>3:
                    print("Detected Deteriorating Accuracy... Stopping Training Prematurely")
                    break
            else:
                det_cnt=0
        print(f"Epoch:{epoch}. Accuracy:{accuracy*100},Precision:{precision*100},Recall:{recall*100},F1 Score:{f1_score*100}")
        plot_epochs.append(epoch)
        acc.append(accuracy)
        prec.append(precision)
        recalls.append(recall)
        f1.append(f1_score)
```
